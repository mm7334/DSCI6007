```
df = pd.read_csv('sample_geo.txt', sep='\t', na_values= 'NaN')
#df = pd.read_csv('sample_geo.txt', sep='\t',header=None)

df = df.dropna()



#df['ADDR_PCT_CD'] = df['ADDR_PCT_CD'].fillna(0)
#df['ADDR_PCT_CD'] = df['ADDR_PCT_CD'].astype('int64')

#df.tail()
df.head()
```

```
fig = plt.figure(figsize = (12, 9))

m = Basemap(projection = 'mill', 
            llcrnrlat = 20,
            urcrnrlat = 55,
            llcrnrlon = -130,
            urcrnrlon = -60,
            resolution = 'c'
            )

m.drawcoastlines(color="black",linewidth=1.2)

m.drawparallels(np.arange(20,60,5), labels = [True, False, False, False])

m.drawmeridians(np.arange(-180,180,30), labels = [0,0,0,1])

lat_y = df['Latitude'].tolist()
lon_x = df['Longitude'].tolist()

colors = 'red'

m.scatter(lon_x, lat_y, latlon = True, s = 50, c = colors, marker = 'o', alpha = 1, edgecolor = 'k', linewidth = 1, zorder = 2)

plt.title('Basemap of geo_data')

plt.show
```

```
fig = plt.figure(figsize = (12, 9))

m = Basemap(projection = 'mill', 
            llcrnrlat = -90,
            urcrnrlat = 90,
            llcrnrlon = -180,
            urcrnrlon = 180,
            resolution = 'c'
            )

m.drawcoastlines(color="black",linewidth=1.2)

m.drawparallels(np.arange(-90,90,10), labels = [True, False, False, False])

m.drawmeridians(np.arange(-180,180,30), labels = [0,0,0,1])

lat_y = df['Latitude'].tolist()
lon_x = df['Longitude'].tolist()

colors = 'red'

m.scatter(lon_x, lat_y, latlon = True, s = 50, c = colors, marker = 'o', alpha = 1, edgecolor = 'k', linewidth = 1, zorder = 2)

plt.title('Basemap of geo_data')

plt.show
```
